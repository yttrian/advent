```kotlin
import kotlin.io.path.Path
import kotlin.io.path.readLines

val input = Path("Day01.txt").readLines()
```

Part 1

```kotlin
fun rotate(current: Int, turn: Int): Int {
    val next = (current + turn).rem(100)
    return if (next >= 0) next else (100 + next)
}
```

```kotlin
fun parseTurn(instruction: String): Int {
    val direction = when (instruction[0]) {
        'L' -> -1
        'R' -> 1
        else -> error("Unknown direction!")
    }
    return instruction.drop(1).toInt() * direction
}
```

```kotlin
input.runningFold(50) { current, instruction ->
    val turn = parseTurn(instruction)
    rotate(current, turn)
}.count { it == 0 }
```

Part 2

```kotlin
fun rotateSteps(current: Int, turn: Int) = buildList {
    val step = turn.compareTo(0)
    var current = current

    repeat(turn.absoluteValue) {
        current = rotate(current, step)
        add(current)
    }
}
```

```kotlin
input.runningFold(listOf(50)) { steps, instruction ->
    val turn = parseTurn(instruction)
    rotateSteps(steps.last(), turn)
}.flatten().count { it == 0 }
```
