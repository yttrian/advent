```kotlin
import kotlin.io.path.Path
import kotlin.io.path.readText

val input = Path("Day02.txt").readText().trim()
```

Part 1

```kotlin
val ranges = input.split(',').map { range ->
    val (left, right) = range.split('-')
    left.toLong()..right.toLong()
}
ranges
```

```kotlin
fun isRepeatedTwice(value: Long): Boolean {
    val text = value.toString()

    // It's okay if it rounds wrong
    val half = text.length / 2

    return text.take(half) == text.drop(half)
}
```

```kotlin
ranges.sumOf { range ->
    range.filter { isRepeatedTwice(it) }.sum()
}
```

Part 2

```kotlin
fun String.progressive() = indices.map { take(it + 1) }
```

```kotlin
fun isRepeating(value: Long): Boolean {
    val text = value.toString()

    // Could be more efficient but eh
    val prefixes = text.progressive().dropLast(1)

    return prefixes.any { prefix ->
        // Conveniently the integer math works out
        prefix.repeat(text.length / prefix.length) == text
    }
}
```

```kotlin
ranges.sumOf { range ->
    range.filter { isRepeating(it) }.sum()
}
```
