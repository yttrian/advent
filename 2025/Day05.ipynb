```kotlin
import kotlin.io.path.Path
import kotlin.io.path.readLines
import kotlin.io.path.readText

val input = Path("Day05.txt").readLines()
```

```kotlin
val ranges = input.takeWhile { it.isNotEmpty() }.map {
    val (from, to) = it.split('-')
    from.toLong()..to.toLong()
}

val ingredients = input.dropWhile { it.isNotEmpty() }.drop(1).map { it.toLong() }
```

```kotlin
// Part 1
ingredients.count { ingredient -> ranges.any { it.contains(ingredient) }}
```

```kotlin
// Part 2
fun subsume(a: LongRange, b: LongRange): LongRange? {
    return if (a.start in b || a.endInclusive in b) {
        min(a.start, b.start)..max(a.endInclusive, b.endInclusive)
    } else {
        null
    }
}

fun reduce(ranges: List<LongRange>): List<LongRange> {
    val set = ranges.toMutableSet()

    for (range in ranges) {
        var accumulator = range

        val used = buildList {
            for (candidate in set) {
                val merge = subsume(accumulator, candidate)

                if (merge !== null) {
                    accumulator = merge
                    add(candidate)
                }
            }
        }

        set.removeAll(used)
        set.add(accumulator)
    }

    return set.toList()
}

reduce(ranges).sumOf { it.endInclusive - it.start + 1 }
```
