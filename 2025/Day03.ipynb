```kotlin
import kotlin.io.path.Path
import kotlin.io.path.readLines
import kotlin.io.path.readText

val input = Path("Day03.txt").readLines()
```

```kotlin
val banks = input.map { bank ->
    bank.map { battery ->
        battery.toString().toInt()
    }
}
```

```kotlin
fun max(bank: List<Int>, targetLength: Int): Long {
    tailrec fun compute(bank: List<Int>, current: String = ""): String {
        val currentLength = current.length

        if (currentLength == targetLength) {
            return current
        }

        val candidate = bank.dropLast(targetLength - currentLength - 1).max()
        val next = current + candidate

        return compute(bank.drop(bank.indexOf(candidate) + 1), next)
    }

    return compute(bank).toLong()
}
```

```kotlin
// Part 1
banks.sumOf { max(it, 2) }
```

```kotlin
// Part 2
banks.sumOf { max(it, 12) }
```
