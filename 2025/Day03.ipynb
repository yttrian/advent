```kotlin
import kotlin.io.path.Path
import kotlin.io.path.readLines
import kotlin.io.path.readText

val input = Path("Day03.txt").readLines()
```

```kotlin
val banks = input.map { bank ->
    bank.map { battery ->
        battery.toString().toInt()
    }
}
```

```kotlin
tailrec fun max(bank: List<Int>, targetLength: Int, current: Long? = null): Long {
    val currentLength = current?.toString()?.length ?: 0

    if (currentLength == targetLength) {
        return current!!
    }

    val candidate = bank.windowed(targetLength - currentLength).maxOf { it.first() }
    val next = (current ?: 0) * 10 + candidate

    return max(bank.drop(bank.indexOf(candidate) + 1), targetLength, next)
}
```

```kotlin
banks.sumOf { max(it, 2) }
```

```kotlin
banks.sumOf { max(it, 12) }
```
